```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#The empty parking spot image
empty_lot = "empty_lot.jpg"
car_in_lot = "car_in_lot_2.jpg"
img = cv2.imread(empty_lot,cv2.IMREAD_COLOR)
box_img = img.copy()
```

```python
#Draw 3 boxes around the parking spots 
img_rect = cv2.rectangle(box_img,(270,480), (450,200), (0,255,0), 5)
img_rect = cv2.rectangle(box_img,(50,480), (200,200), (0,255,0), 5)
img_rect = cv2.rectangle(box_img,(500,480), (700,200), (0,255,0), 5)
```

```python
plt.imshow(img)
```

```python
plt.imshow(img_rect)
```

```python
#Empty spot images 
spot1_coords = [[200,480],[50,200]]
spot2_coords = [[200,480],[270,450]]
spot3_coords = [[200,480],[500,700]]
all_spots = [spot1_coords,spot2_coords,spot3_coords]
spot1 = img[200:480, 50:200] 
spot2 = img[200:480, 270:450] 
spot3 = img[200:480, 500:700] 
empty_spots = [spot1,spot2,spot3]
```

```python
car_in_lot = "car_in_lot.jpg"
new_img = cv2.imread(car_in_lot,cv2.IMREAD_COLOR)
color_new_img = new_img.copy()
for i in range(3): 
    img_orig = empty_spots[i]
    cropped = new_img[all_spots[i][0][0]:all_spots[i][0][1],all_spots[i][1][0]:all_spots[i][1][1]]
    masked = (cropped - img_orig).flatten()
    non_zero = cv2.countNonZero(masked)
    num_pixels = img_orig.shape[0]*img_orig.shape[1]*img_orig.shape[2]
    if non_zero / num_pixels > 0.5: 
        print('Spot ',i,' filled')
        cv2.rectangle(color_new_img,(all_spots[i][1][0],all_spots[i][0][1]), (all_spots[i][1][1],all_spots[i][0][0]), (255,0,0), 5)
    else: 
        cv2.rectangle(color_new_img,(all_spots[i][1][0],all_spots[i][0][1]), (all_spots[i][1][1],all_spots[i][0][0]), (0,255,0), 5)
```

```python
plt.imshow(color_new_img)
```
